# Introduction: Présentation de la base de données et de notre problématique

La base de données sur laquelle nous avons travaillé contient des informations sur des experience de Speeddating qui se sont déroulées entre 2002 et 2004. Le principe est simple: chaque participant a eu des rendez-vous de 4 minutes avec tous les autres participants du sexe opposées. A la fin des 4 minutes, les participants ont repondu si ils souhaitaient revoir leur partenaire ou non.  On leur a demandé de noter 6 attributs chez leur partenaire: l'attraction, la sincerité, l'intelligence, l'ambition, les interets commun, et l'humour.  

La base de données contient aussi des informations sur les participants qui ont été recoltés tout au long de l'experience a l'aide de questionnaires. Les informations concernent le style de vie, ce que chacun cherche chez un partenaire, le metier, les croyances, l'age, etc...


Notre problématique est la suivante: Peut-on prédire si deux personnes vont vouloir se revoir à la fin de leur rendez-vous? 

Pour répondre a cette problématique, nous avons trouvé interressant d'adopté et de comparer deux modèles: un modèle supervisé et un modèle non supervisé. Dans le modèle supervisé, notre variable d'output est la variable qui correspond à la réponse à la question suivante: "Souhaitez vous revoir cette personne?". Nos variable d'inputs  correspondent aux reponses a diverses questions qui resument les personnalités et les profils des individus, ainsi que  l'importance que l'individu apporte aux différents attributs citéplus haut. Ces notes  correspondent à la réponse à la question suivante: "You have 100 points to distribute among the following attributes -- give more points to those attributes that are more important in a potential date, and fewer points to those attributes that are less important in a potential date".

# Partie II: implementations des fonctions 

Dans cette partie, nous avons implementé les fonctions qui vous nous etres utiles dans la suite.
Nous avons tout d'abord implémenté quelques fonctions qui permettent de couper le dataframe en une partie (Xtest, Ytest) et une partie (Xtrain,Ytrain). 

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
def split(frac,df):
    train=df.sample(frac=frac,random_state=200)
    test=df.drop(train.index)
    return(train,test)
def XY(frac,df):
    train,test=split(frac,df)
    Ytrain=train.dec.values
    Ytest=test.dec.values
    Xtrain=train.drop("dec",1).values
    Xtest=test.drop("dec",1).values
    return(Xtrain,Xtest,Ytrain,Ytest)
```

Voila quelques fonctions qui nous renvoie differents score pour un modèle donné. La fonction score_f1 nous renvoie le f1score et la fonction score_roc nous renvoie l'air sous la courbe roc. Les scores sont calculés en utilisant un k cross validation avec k=10, et en faisant la moyenne des scores obtenus.

```python
def score_f1(clf,X,Y):
    kf = KFold(n_splits=10,shuffle=True)
    scores = cross_val_score(clf, X,Y, cv=kf,scoring="f1")
    return(np.mean(scores))

def score_roc(clf,X,Y):
    kf = KFold(n_splits=10,shuffle=True)
    scores = cross_val_score(clf, X,Y, cv=kf,scoring="roc_auc")
    return(np.mean(scores))
    
    
```

Maintenant, nous pouvons entrainer le random forest. Interressons nous à l'efficacité de la prédiction lorsque nous ne changeons aucun paramètre dans le random forest, et lorsque nous demandons au modèles de predire Y lorsque nous lui donnons respectivemetn Xtrain et Xtest.

```python
l=[]
m=[]
for i in range (20):
    Xtrain,Xtest,Ytrain,Ytest=XY(0.8,df4)
    clf=RandomForestClassifier()
    clf.fit(Xtrain,Ytrain)
    Ypred=clf.predict(Xtest)
    l.append(f1_score(Ytest,Ypred))
    m.append(clf.score(Xtrain,Ytrain))
```

Dans la prochaine partie, nous allons donc essayer de changer les paramètres de l'arbre de décision de manière à optimiser la prédiction en évitant le overfitting. 

# Partie IV: Calibrage des hyperparamètres

Dans cette partie, nous allons essayer de trouver les paramètres du randomforest optimaux tour d'abord à la main, et ensuite a l'aide d'un gridsearch. Il faut garder en tête que les classes de notre modèle sont déséquilibrées. Nous avons donc choisi de s'intérresser à la courbe ROC pour évaluer l'efficacité de nos modèles plutot que la matrice de confusion.

```python
from sklearn.metrics import roc_curve, auc, confusion_matrix
```

Interressons nous tout d'abord à l'influence de n_estimators. Nous avons fait varier n_estimators de 5 à 500, 30 par 30.

```python
Xtrain,Xtest,Ytrain,Ytest=XY(0.8,df4)
l=np.arange(5,500,30)

for i in l:
    clf=RandomForestClassifier(n_estimators=i)
    clf.fit(Xtrain,Ytrain)
    probas = clf.predict_proba(Xtest)
# probas est une matrice de deux colonnes avec la probabilités d'appartenance à chaque classe
    fpr, tpr, thresholds = roc_curve(Ytest, probas[:, 1])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label="ROC curve n= {} (area ={})".format(i,round(roc_auc,3)))

plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right", fontsize="x-small")
```

Nous remarquons qu'à partir d'environ 130 arbres, augmenter le nombres d'arbres dans notre random forest n'augmente pas consirablement l'efficacité de la prédiction (l'air sous la courbe). Pour ne pas perdre en vitesse, nous avons retenu: n_estimators=130.

Faisons la même chose en faisant varier max_depth de 1 à 25. Cette fois ci, nous nous interressons à l'evolution de l'air sous la courbe lorsque l'on prédit sur Xtrain et Xtest respectivement.

```python
Xtrain,Xtest,Ytrain,Ytest=XY(0.8,df4)
l=np.arange(1,25)
m=[]
n=[]

for i in l:
    clf=RandomForestClassifier(n_estimators=130, max_depth=i)
    clf.fit(Xtrain,Ytrain)
    probas1 = clf.predict_proba(Xtest)
    probas2=clf.predict_proba(Xtrain)
    fpr1, tpr1, thresholds1 = roc_curve(Ytest, probas1[:, 1])
    fpr2, tpr2, thresholds2 = roc_curve(Ytrain, probas2[:, 1])
    m.append(auc(fpr1, tpr1))
    n.append(auc(fpr2,tpr2))

plt.plot(m, label="Prédiction sur Xtest")
plt.plot(n,label="Prédiction sur Xtrain")
plt.ylabel('Air sous la courbe')
plt.xlabel('maximum depth')
plt.title("Air sous la courbe en fonction de la profondeur maximal des arbres de décisions")
plt.legend(loc="lower right", fontsize="x-small")
```

Nous remarquons qu'à partir d'une profondeur de plus de 10, le modèle apprend parfaitement Xtrain, ce qui peut être dangereux parceque nous ne voulons pas que le modèle apprenne du bruit. Mais les performances  du modèle sur Xtest ne semblent pas diminuer , il semble donc inutile de restreindre ce paramètre.

Ensuite nous avons fait varier min_samples_leaf de 1 a 25:

```python
Xtrain,Xtest,Ytrain,Ytest=XY(0.8,df4)
l=np.arange(1,25)
m=[]
n=[]

for i in l:
    clf=RandomForestClassifier(n_estimators=130, min_samples_leaf=i)
    clf.fit(Xtrain,Ytrain)
    probas1 = clf.predict_proba(Xtest)
    probas2=clf.predict_proba(Xtrain)
    fpr1, tpr1, thresholds1 = roc_curve(Ytest, probas1[:, 1])
    fpr2, tpr2, thresholds2 = roc_curve(Ytrain, probas2[:, 1])
    m.append(auc(fpr1, tpr1))
    n.append(auc(fpr2,tpr2))

plt.plot(m, label="Prédiction sur Xtest")
plt.plot(n,label="Prédiction sur Xtrain")
plt.ylabel('Air sous la courbe')
plt.xlabel('min_samples_leaf')
plt.title("Air sous la courbe en fonction du nombre minimum d'echantillon par feuilles autorisé ")
plt.legend(loc="lower right", fontsize="x-small")
```

Il semble qu'il n'y ai pas de overfitting. Contraindre le nombre minimum d'échantillons par feuilles ne fait que réduire l'efficacité dans les deux cas. Interessons nous maintenant à l'influence du nombre maximum de variables autorisées dans chaque arbres de décisions. 

```python
Xtrain,Xtest,Ytrain,Ytest=XY(0.8,df4)
l=np.arange(1,25)
m=[]
n=[]

for i in l:
    clf=RandomForestClassifier(n_estimators=130, max_features=i)
    clf.fit(Xtrain,Ytrain)
    probas1 = clf.predict_proba(Xtest)
    probas2=clf.predict_proba(Xtrain)
    fpr1, tpr1, thresholds1 = roc_curve(Ytest, probas1[:, 1])
    fpr2, tpr2, thresholds2 = roc_curve(Ytrain, probas2[:, 1])
    m.append(auc(fpr1, tpr1))
    n.append(auc(fpr2,tpr2))

plt.plot(m, label="Prédiction sur Xtest")
plt.plot(n,label="Prédiction sur Xtrain")
plt.ylabel('Air sous la courbe')
plt.xlabel('max_features')
plt.title("Air sous la courbe en fonction du nombre maximum de variables autorisées dans chaques arbres de décisions ")
plt.legend(loc="lower right", fontsize="x-small")
```

Il semble qu'a partie de 7 variables, le score diminue, nous avons donc choisi max_features=7.

Maintenant que nous avons essayé de calibrer les premiers hyper-parametres à la main, nous avons entrepris de faire une recherche automatisée à l'aide d'un gridsearch pour les parametres restants (max_leaf_nodes et class_weight)

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid= {"max_leaf_nodes":[500,800,900],"class_weight":[{0:1,1:1},{0:85,1:1},{0:0.8,1:1},{0:0.9,1:1}]}
```

```python
Xtrain,Xtest,Ytrain,Ytest=XY(0.8,df4)
clf1=RandomForestClassifier(n_estimators=130,min_samples_split=5,max_features=7)
clf=GridSearchCV(clf1,param_grid)
clf.fit(Xtrain,Ytrain)
```

```python
print(clf.best_params_)
```

Au final, les hyper-paramètres optimaux semblent etres:

n_estimators=130,min_samples_split=5,class_weight={0:0.85,1:1},max_leaf_nodes=800,max_features=7

    

En prenant ces paramètres, interressons nous aux scores de notre modèles:

```python
Xtrain,Xtest,Ytrain,Ytest=XY(0.8,df4)
clf=RandomForestClassifier(n_estimators=130,min_samples_split=5,class_weight={0:0.85,1:1},
                           max_leaf_nodes=800,max_features=7)
X=df4.drop("dec",1).values
Y=df4.dec.values
```

```python
print(score_f1(clf,X,Y),score_roc(clf,X,Y))
```

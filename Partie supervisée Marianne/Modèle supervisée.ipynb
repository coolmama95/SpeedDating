## Partie I: choix des variables et nettoyage de la bas de données

Dans cette partie, nous choisissons les variables qui nous interressent, et nous conservons le dataframe nettoyé contenant toutes ses variabes pour l'exploiter dans la seconde partie. 

Notre but final est de construire un modèle supervisée qui puissent prédire si deux personnes qui se sont rencontrées lors d'une séance de SpeedDating souhaitent se revoir.

La décision final est contenu dans la variable dec (1=yes et 0=no) et correspond à la réponse à la question suivante: "Souhaitez vous revoir cette personne?". la variable dec est donc la variable à prédire et est par consequent notre variable d'output.

Les variables d'input que nous avons choisis correspondent tout d'abord à l'importance que l'individu apporte a different critères: attr1_1 (Attractive), sinc1_1 (Sincere), intel1_1 (Intelligent),fun1_1 (Fun ),amb1_1 (Ambitious), shar1_1 (Shared Interests/Hobbies)  et correspondent à la réponse à la question suivante: "You have 100 points to distribute among the following attributes -- give more points to those attributes that are more important in a potential date, and fewer points to those attributes that are less important in a potential date"

Les notes attribuées sont contenus dans les variables: attr, sinc, intel, fun, amb, shar.
La variable gender contient le genre de la personne.


La totalité des variables qui nous intérressent et leurs valeurs sont contenus dans le dataframe df2. Nous avons procédés à un nettoyage du dataframe (Nous avons enlever les Nan et enlever les individus dont la sommes des réponses ne faisait pas 100)

```python
list_qual=["attr","sinc","intel","fun","amb","shar","dec", "gender"]
list_var=[i for i in df.columns if ((any(x in i for x in list_qual)&("1_1" in i))| (i in list_qual))]
df2=df[(df.wave<6)|(df.wave>9)]
```

```python
df2=pd.concat([df2[[i]] for i in list_var],axis=1).dropna()
```

```python
df2["somme"]=df2.attr1_1+ df2.sinc1_1+df2.intel1_1+df2.amb1_1+df2.shar1_1 +df2.fun1_1
df2.somme.unique()
df2=df2[(df2.somme<110)&(df2.somme>95)].drop("somme",1)
```

# Partie II: Premier Random Forest

Maintenant que nous avons un dataframe exploitable, nous allons essayer dans cette partie de faire un premier modèle supervisée à l'aide d'un RandomForest.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, f1_score
from sklearn.tree import export_graphviz
```

Nous avons implémenté quelques fonctions qui permettent de couper le dataframe en une partie (Xtest, Ytest) et une partie (Xtrain,Ytrain). 

```python
def split(frac,df):
    train=df.sample(frac=frac,random_state=200)
    test=df.drop(train.index)
    return(train,test)
def XY(frac,df):
    train,test=split(frac,df)
    Ytrain=train.dec.values
    Ytest=test.dec.values
    Xtrain=train.drop("dec",1).values
    Xtest=test.drop("dec",1).values
    return(Xtrain,Xtest,Ytrain,Ytest)
def n_estimators(n,frac,df):
    clf=RandomForestClassifier(n_estimators=n)
    train,test=split(frac,df)
    clf.fit(XY(train,test)[0], XY(train,test)[2])
    Ypred=clf.predict(XY(train,test)[1])
    return(f1_score(XY(train,test)[3],Ypred ))
```

```python
Xtrain,Xtest,Ytrain,Ytest=XY(0.8,df2)
```

Maintenant, nous pouvons entrainer le random forest. Interressons nous à l'efficacité de la prédiction lorsque nous ne changeons aucun paramètre dans le random forest.

```python
l=[]
m=[]
for i in range (50):
    Xtrain,Xtest,Ytrain,Ytest=XY(0.8,df2)
    clf=RandomForestClassifier()
    clf.fit(Xtrain,Ytrain)
    Ypred=clf.predict(Xtest)
    l.append(f1_score(Ytest,Ypred))
    m.append(clf.score(Xtrain,Ytrain))
```

```python
print(np.mean(l),np.mean(m))
```

En moyenne, le f1 score de la prédiction sur Xtrain est de 0.98, alors qu'il est de 0.73 pour Xtest. Il est bien plus élevé sur Xtrain, signe de overfitting. # Partie III: Parameter fitting à la main


```python
from sklearn.metrics import roc_curve, auc
```

Dans cette partie, nous allons essayer de trouver les paramètres du randomforest optimaux à la main. Il faut garder en tête que les classes de notre modèle sont déséquilibrées. Nous avons donc choisi de s'intérresser à la courbe ROC pour évaluer l'efficacité de nos modèles plutot que la matrice de confusion.

Nous allons  essayer de jouer sur les paramètres suivant:  n_estimators, max_depth,  min_samples_leaf, max_features.

Interressons nous tout d'abord à l'influence de n_estimators. Nous avons fait varier n_estimators de 5 a 500, 30 par 30.

```python
from sklearn.metrics import roc_curve, auc

Xtrain,Xtest,Ytrain,Ytest=XY(0.8,df2)
l=np.arange(5,500,30)

for i in l:
    clf=RandomForestClassifier(n_estimators=i)
    clf.fit(Xtrain,Ytrain)
    probas = clf.predict_proba(Xtest)
# probas est une matrice de deux colonnes avec la proabilités d'appartenance à chaque classe
    fpr, tpr, thresholds = roc_curve(Ytest, probas[:, 1])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label="ROC curve n= {} (area ={})".format(i,round(roc_auc,3)))

plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right", fontsize="x-small")

```

Nous remarquons qu'à partir d'environ 100 arbres, augmenter le nombres d'arbres dans notre random forest n'augmente pas consirablement l'efficacité de la prédiction (l'air sous la courbe). Pour ne pas perdre en vitesse, nous avons retenu: n_estimators=100.

Faisons la même chose en faisant varier max_depth de 1 à 25. Cette fois ci, nous nous interressons à l'evolution de l'air sous la courbe lorsque l'on prédit sur Xtrain et Xtest respectivement.

```python
Xtrain,Xtest,Ytrain,Ytest=XY(0.8,df2)
l=np.arange(1,25)
m=[]
n=[]

for i in l:
    clf=RandomForestClassifier(n_estimators=100, max_depth=i)
    clf.fit(Xtrain,Ytrain)
    probas1 = clf.predict_proba(Xtest)
    probas2=clf.predict_proba(Xtrain)
    fpr1, tpr1, thresholds1 = roc_curve(Ytest, probas1[:, 1])
    fpr2, tpr2, thresholds2 = roc_curve(Ytrain, probas2[:, 1])
    m.append(auc(fpr1, tpr1))
    n.append(auc(fpr2,tpr2))

plt.plot(m, label="Prédiction sur Xtest")
plt.plot(n,label="Prédiction sur Xtrain")
plt.ylabel('Air sous la courbe')
plt.xlabel('maximum depth')
plt.title("Air sous la courbe en fonction de la profondeur maximal des arbres de décisions")
plt.legend(loc="lower right", fontsize="x-small")


```

Nous remarquons qu'à partir d'une profondeur de plus de 15, le modèle apprend parfaitement Xtrain, ce qui peut être dangereux parceque nous ne voulons pas que le modèle apprenne du bruit. Les performances du modèles sur Xtest stagnent à partir de max_depth=12, nous avons choisi de retenir cette valeur.

```python
Xtrain,Xtest,Ytrain,Ytest=XY(0.8,df2)
l=np.arange(1,25)
m=[]
n=[]

for i in l:
    clf=RandomForestClassifier(n_estimators=100, max_depth=12, min_samples_leaf=i)
    clf.fit(Xtrain,Ytrain)
    probas1 = clf.predict_proba(Xtest)
    probas2=clf.predict_proba(Xtrain)
    fpr1, tpr1, thresholds1 = roc_curve(Ytest, probas1[:, 1])
    fpr2, tpr2, thresholds2 = roc_curve(Ytrain, probas2[:, 1])
    m.append(auc(fpr1, tpr1))
    n.append(auc(fpr2,tpr2))

plt.plot(m, label="Prédiction sur Xtest")
plt.plot(n,label="Prédiction sur Xtrain")
plt.ylabel('Air sous la courbe')
plt.xlabel('min_samples_leaf')
plt.title("Air sous la courbe en fonction du nombre minimum d'echantillon par feuilles autorisé ")
plt.legend(loc="lower right", fontsize="x-small")
```

Il semble qu'il n'y ai pas de overfitting. Contraindre le nombre minimum d'échantillons par feuilles ne fait que réduire l'efficacité dans les deux cas. 

Interessons nous maintenant à l'influence du nombre maximum de variables autorisées dans chaque arbres de décisions. Il y a 14 variables, nous avons par conséquent fait varier le paramètre de 1 à 14.

```python
Xtrain,Xtest,Ytrain,Ytest=XY(0.8,df2)
l=np.arange(1,13)
m=[]
n=[]

for i in l:
    clf=RandomForestClassifier(n_estimators=100, max_depth=12, max_features=i)
    clf.fit(Xtrain,Ytrain)
    probas1 = clf.predict_proba(Xtest)
    probas2=clf.predict_proba(Xtrain)
    fpr1, tpr1, thresholds1 = roc_curve(Ytest, probas1[:, 1])
    fpr2, tpr2, thresholds2 = roc_curve(Ytrain, probas2[:, 1])
    m.append(auc(fpr1, tpr1))
    n.append(auc(fpr2,tpr2))

plt.plot(m, label="Prédiction sur Xtest")
plt.plot(n,label="Prédiction sur Xtrain")
plt.ylabel('Air sous la courbe')
plt.xlabel('max_features')
plt.title("Air sous la courbe en fonction du nombre maximum de variables autorisées dans chaques arbres de décisions ")
plt.legend(loc="lower right", fontsize="x-small")
```

La encore, il semble qu'il n'y ai pas d'overfitting, inutile de restreindre ce paramètre.
